```python
#impport pandas library
import pandas as pd

#Initialize url value with one given in assignment
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

print(url)
lst = pd.read_html(url)

#Print the total number of tables found of url
print("Total tables found on the webpage: " + str(len(lst)))

#Create Dataframe of the first table
df = pd.DataFrame(lst[0])

#print("Type of df2: " + str(type(df)))

#df.shape

#List the first 10 rows of datafram
df.head(10)
```

### Let's remove records where Borough is not assigned

```python
df = df[df.Borough != 'Not assigned'].reset_index(drop=True)
df
```

### Get the shape (number of rows and columns) of the dataframe

```python
df.shape
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df['Borough'].unique()),df.shape[0]))
```

## Code below this cell is for getting longitude and latitude details from csv file

```python
toronto_location = pd.read_csv('http://cocl.us/Geospatial_data')
print(toronto_location)
```

```python
toronto_location.rename(columns = {'Postal Code':'Postal code'}, inplace = True)
toronto_location
```

```python
df_toronto_neighborhoud = pd.merge(df, toronto_location, on='Postal code', how='inner')
df_toronto_neighborhoud
```

### Keep only those boroughs in which word 'Toronto' is present

```python
df_toronto_boroughs = df_toronto_neighborhoud[df_toronto_neighborhoud.Borough.str.contains('Toronto',case=False)].reset_index(drop=True)
df_toronto_boroughs
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="tr_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, lng, label in zip(df_toronto_boroughs['Latitude'], df_toronto_boroughs['Longitude'], df_toronto_boroughs['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
